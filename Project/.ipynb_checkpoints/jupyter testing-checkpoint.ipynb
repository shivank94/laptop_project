```python
from flask import Flask, render_template, request
import pickle
import pandas as pd
```

```python
with open('encoders.pkl', 'rb') as f1:
    encoders = pickle.load(f1)
```

```python
with open('test_df.pkl', 'rb') as f2:
    test_df = pickle.load(f2)
```

```python
with open('xgboost_model.pkl', 'rb') as f3:
    model = pickle.load(f3)
```

```python
default_input = {
    'company': 'Dell',
    'product': 'Inspiron',
    'typename': 'Notebook',
    'inches': 15.6,
    'cpu': 'Intel Core i5 7200U',
    'ram(GB)': 8,
    'gpu': 'Intel UHD Graphics 620',
    'opsys': 'Windows 10',
    'weight(kg)': 2.0,
    'resolution': '1920x1080',
    'screentype': 'IPS',
    'touchscreen': 0,
    'cpu_freq(GHz)': 2.5,
    'memory_1_sto(GB)': 256,
    'memory_1_type': 'SSD',
    'memory_2_sto(GB)': 0,
    'memory_2_type': 'None',
    'cpu_brand': 'Intel',
    'gpu_brand': 'Intel'
}

feature_order = ['company', 'product', 'typename', 'inches', 'cpu', 'ram(GB)', 'gpu',
       'opsys', 'weight(kg)', 'resolution', 'screentype',
       'touchscreen', 'cpu_freq(GHz)', 'memory_1_sto(GB)', 'memory_1_type',
       'memory_2_sto(GB)', 'memory_2_type', 'cpu_brand', 'gpu_brand']
```

```python
input_data = {
    'company': 'Dell',
    'typename': 'Notebook',
    'inches': 15.6,
    'cpu': 'Intel Core i7 7200U',
    'ram(GB)': 8,
    'gpu': 'Intel UHD Graphics 620',
    'opsys': 'Windows 10',
    'resolution': '1920x1080',
    'touchscreen': 0,
    'memory_1_sto(GB)': 256,
    'memory_1_type': 'SSD',
    'cpu_brand': 'Intel',
    'gpu_brand': 'Intel'
}
```

```python
# Fill missing inputs
complete_input = default_input.copy()
complete_input.update(input_data)

# Encode categorical columns
for col in feature_order:
    if col in encoders:
        le = encoders[col]
        val = complete_input[col]
        if val in le.classes_:
           complete_input[col] = le.transform([val])[0]
        else:
            # handle unknown category
            complete_input[col] = -1  # or use most common class, etc.

input_df = pd.DataFrame([[complete_input[col] for col in feature_order]], columns=feature_order)
```

```python
# Start with a copy of the default input
complete_input = default_input.copy()

# Update with user-provided values
complete_input.update(input_data)

# Encode categorical columns (including values from default_input)
for col in feature_order:
    if col in encoders:
        le = encoders[col]
        val = complete_input[col]
        if val in le.classes_:
            complete_input[col] = le.transform([val])[0]
        else:
            # Handle unknown category
            complete_input[col] = -1  # or use a fallback like most frequent class
    else:
        # Ensure non-categorical values remain unchanged
        complete_input[col] = complete_input[col]

# Convert to DataFrame
input_df2 = pd.DataFrame([[complete_input[col] for col in feature_order]], columns=feature_order)
```
